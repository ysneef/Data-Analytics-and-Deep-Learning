2. Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1) như sau: 

a. Từ 0 đến dưới 5 = kém (ký hiệu “k”)  

b. Từ 5 đến dưới 7 = trung bình (ký hiệu “tb”) 

c. Từ 7 đến dưới 8 = khá (ký hiệu “k”) 

d. Từ 8 trở lên = giỏi (ký hiệu “g”) 

```python
df['phanlopt1'] = np.where(df['T1'] < 5, 'k',
                           np.where(df['T1'] < 7, 'tb',
                                    np.where(df['T1'] < 8, 'kh', 'g')))

```

```python
freq = df['phanlopt1'].value_counts()
print(freq)
```

4. Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến phanlopt1

```python
for group in df['phanlopt1'].unique():
    plt.plot(df[df['phanlopt1'] == group]['T1'], label=group)

plt.xlabel('Thứ tự')
plt.ylabel('Giá trị T1')
plt.title('Biểu đồ đường Multiple Line cho biến T1 theo phân loại')
plt.legend()
plt.show()
```

5. Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanlopt1

Phần 5: Mô tả dữ liệu và khảo sát dạng phân phối

1. Hãy mô tả và khảo sát phân phối cho biến T1 

```python
print("Mô tả thống kê biến T1:")
print(df['T1'].describe())

# Vẽ boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(y='T1', data=df)
plt.title('Boxplot của T1')
plt.show()
```

```python
Q1 = df['T1'].quantile(0.25)
Q3 = df['T1'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['T1'] < lower_bound) | (df['T1'] > upper_bound)]
print(f"Min: {df['T1'].min()}")
print(f"Q1: {Q1}")
print(f"Median: {df['T1'].median()}")
print(f"Q3: {Q3}")
print(f"Max: {df['T1'].max()}")
print(f"IQR: {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Outliers: \n{outliers}")
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['T1'], kde=True)
plt.title('Histogram của T1')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
stats.probplot(df['T1'], dist="norm", plot=plt)
plt.title('QQ-plot của T1')
plt.show()

```

Nhận xét:
- Dựa vào boxplot, ta thấy có một số giá trị ngoại lai (outliers).
- Histogram cho thấy phân phối của T1 có vẻ hơi lệch phải.
- QQ-plot cho thấy dữ liệu không hoàn toàn tuân theo phân phối chuẩn.

2. Hãy mô tả và khảo sát phân phối cho biến T1 trên từng nhóm phân lớp (phanlopT1)

```python
bins = [0, 5, 7, 9, 10]
labels = ['Yếu', 'Trung bình', 'Khá', 'Giỏi']
df['phanlopT1'] = pd.cut(df['T1'], bins=bins, labels=labels, right=False)


# Boxplot theo nhóm phanlopT1
plt.figure(figsize=(10, 6))
sns.boxplot(x='phanlopT1', y='T1', data=df)
plt.title('Boxplot của T1 theo nhóm phanlopT1')
plt.show()
```

```python
# QQ-plot theo nhóm phanlopT1
for category in df['phanlopT1'].unique():
    plt.figure(figsize=(8, 6))
    stats.probplot(df[df['phanlopT1'] == category]['T1'], dist="norm", plot=plt)
    plt.title(f'QQ-plot của T1 nhóm {category}')
    plt.show()
```

3. Hãy khảo sát tương quan giữa biến DH1 theo biến T1 

```python
correlation = df['DH1'].corr(df['T1'])
print(f"\nTương quan giữa DH1 và T1: {correlation}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x='T1', y='DH1', data=df)
plt.title('Biểu đồ Scatter giữa DH1 và T1')
plt.show()
```

4. Hãy khảo sát tương quan giữa biến DH1 theo biến T1 trên từng nhóm khu vực 

```python
for kv in df['KV'].unique():
    df_kv = df[df['KV'] == kv]
    correlation_kv = df_kv['DH1'].corr(df_kv['T1'])
    print(f"\nTương quan giữa DH1 và T1 ở khu vực {kv}: {correlation_kv}")

    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='T1', y='DH1', data=df_kv)
    plt.title(f'Biểu đồ Scatter giữa DH1 và T1 ở khu vực {kv}')
    plt.show()
```
